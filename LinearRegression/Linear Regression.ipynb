```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16,9]

# reading data
data = pd.read_csv('F:\WorkSpaces\StatisticsPractices\LinearRegression\GDPGermany.csv')
print(data.shape)
data.head()
```

```python
# collecting x and y
X = data['Year']
Y = data['GDP(usd)']

# mean x (since Linear Regression)
mean_x = np.mean(X)
mean_y = np.mean(Y)

# calculating regression line (y = ax + b) of GDP per year
# calculating coefficients a and b
m = len(X)
```

The most commonly-used criterion for the best-fitting line is the line that minimizes the sum of the squared errors of prediction.
The cooefficients a, b are computed with the following formula:

$a = \displaystyle R \frac{\sigma_Y}{\sigma_X}$

where $R$ is corelation between X and Y; $\sigma_Y$ and $\sigma_X$ are standard deviation of X and Y respectively.

$b = \bar{Y} - a\bar{X}$

where $\bar{Y}$ and $\bar{X}$ are mean of Y and X respectively.

------------

Standard deviation can be computed using the following formula: 

$\sigma = \displaystyle\sum\frac{(x-\bar{x})}{(n-1)}$

where $n$ is total number of samples.

Computation can be implemented as follow:


```python
numer = 0
denom = 0

for i in range(m):
    numer += (X[i] - mean_x)*(Y[i] - mean_y)
    denom += (X[i] - mean_x)**2
a = numer/denom
b = mean_y - (a*mean_x)

print(a,b)
```

```python
# plotting
max_x = np.max(X) + 1
min_x = np.min(X) - 1

# interpolating more value
x = np.linspace(min_x, max_x, 1000)
y = a*x + b

plt.plot(x, y, color='r', label = 'Regression Line')
plt.scatter(X, Y, c='b', label = 'Scatter Plot')

plt.xlabel('Years')
plt.ylabel('GDP(usd)')
plt.grid()
plt.xticks(range(X.min(), (X.max()+3), 3), range(X.min(), X.max()+3, 3))
plt.legend()
plt.show()
```

The GDP per capita (in USD) from 1980 - 2018 is stored in a .csv files. This is a simple example of finding a fitting function of a time-series data.
